# **Editing an items**

```python
prod_id = str(input("Enter product id: "))
prod_name=str(input("Enter product name: "))
prod_discount=int(input("Enter discount: "))
prod_expire=str(input("Enter expiry date: "))
prod_mfg=int(input("Enter mfg: "))
prod_price=int(input("Enter product price"))
prod_quantity=int(input("Enter quantity"))

rec[prod_id]['discount']=prod_discount
rec[prod_id]['price']=prod_price
rec[prod_id]['quantity']=prod_quantity
```

# **Purchasing an item**


```python
print("Our inventory have ",end="")
for i in rec:
  print("item id " + i, end=" ")
  print("and item " + rec[i]['product'], end =", ")
item=str(input("Enter your item id : "))
quant=int(input("Enter quantity : "))
name=str(input("Enter your name: "))

total_bill =rec[item]['price']*quant -(rec[item]['price']*rec[item]['discount'])/100

if (rec[item]['quantity']>0):
  if (rec[item]['quantity']<quant):
    print("Sorry {} we have only {} quantity!".format(name,rec[item]['quantity']))
  else:
    out="Hello {} ,thank you so much for purchasing the {} having MFG - {} and Expires at {} .Your billing amount is {}₹ with giving you {} % discount".format(name,rec[item]['product'],rec[item]['mfg'],rec[item]['expire'],total_bill,rec[item]['discount'])
    print("###########################################################################")
    print(out)
else:
  print("Sorry {} , We are out of stock!".format(name))  


rec[item]['quantity']=rec[item]['quantity']-quant
```

# **Refunding an items**

```python
prod=str(input("Enter your item id : "))
print("************** Hello your previous item id and quantity of product {} is {} and {} *****************".format(rec[item]['product'],item,quant))
qua=int(input("Enter how much quantity do you want to refund : "))
nam=str(input("Enter your name: "))

if name==nam:
  if qua<=quant and prod == item:
    rec[prod]['quantity']=rec[prod]['quantity']+qua
    print("Hello {} Your refund is done! Your refund Amount {}₹ has been credited.".format(name,rec[item]['price']*qua -(rec[item]['price']*rec[item]['discount'])/100))
  else:
    print("You are exceeding the quantity nd maybe you are entered incorrect product id")  
else:
  print("Your name is not in the inventroy. Please correct your details.")    
```
